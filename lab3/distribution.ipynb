```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Normal Distribution

```python
#normal distribution

data = np.random.normal(loc=50,scale=10,size=100)
print(data)
plt.hist(data,bins=30)
plt.title("Normal Distribution")
plt.xlabel("value")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```

# Power Law distribution

```python
# power law distribution
data1 = np.random.power(a=5,size=1000)
print(data1)
plt.hist(data1,bins=10,edgecolor='black')
plt.title('power distribution')
plt.xlabel("value")
plt.ylabel("freq")
plt.grid(True)
plt.show()
```

# Distributions

```python
#3. Distributions
from scipy.stats import norm
 
```

```python
np.random.seed(42)
data = pd.DataFrame({
    "age":np.random.randint(20,60,200),
    "salary":np.random.normal(50000,10000,200),
    "department":np.random.choice(["HR","IT","sales","finance"],200)
})
data
```

```python
#histogram of age
plt.hist(data['age'],bins=15,edgecolor="red")
plt.title("histogram of age distribution")
plt.xlabel("values")
plt.ylabel("frequency")
plt.show()
```

```python
#barplot (department counts)
dept_counts =data["department"].value_counts()
dept_counts
```

```python
plt.bar(dept_counts.index,dept_counts.values)
plt.title("employess per department")
plt.xlabel("department")
plt.ylabel("count")
plt.show()
# dept_counts.values
```

# Normalize Distribution (salary)

```python
mean = data["salary"].mean()
std = data["salary"].std()

x = np.linspace(mean-3*std,mean+3*std,200)
y = norm.pdf(x,mean,std)

plt.hist(data["salary"],bins=20,density=True,edgecolor="Green")
plt.plot(x,y,linewidth=2)
plt.title("Salary Distribution with normal Curve")
plt.xlabel("salary")
plt.ylabel("density")
plt.show()
```
