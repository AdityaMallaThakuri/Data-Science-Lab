```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
customers = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4, 5],
    'Name': ['Aditya', 'Nishcal', "Sudip", None, None], 
    'Age': [22, np.nan, 14, np.nan, 28]
})

orders = pd.DataFrame({
    'OrderID': [1, 2, 3, 4],
    'CustomerID': [1, 2, 3,None],
    'Amount': ["300",None,"200","400"]
})

print("\nOriginal Customers Data: ")

print(orders)
print(customers)
```

```python
# remove null values
new_customer = customers.dropna(subset=["Name"])
new_customer
```

```python
new_customer["Age"] = new_customer["Age"].fillna(new_customer['Age'].mean())
new_customer

    
```

```python
orders["Amount"] = pd.to_numeric(orders['Amount'],errors="coerce").fillna(0)
orders
```

```python
merge = pd.merge(new_customer,orders, on = "CustomerID",how="left")
merge
```

```python
merge["Amount_in_thousands"]= merge["Amount"]/1000
merge
```

```python
#normalize age
merge["normalized_age"] = ((merge["Age"]-merge["Age"].min()))/(merge["Age"].max()-merge["Age"].min())
merge
```

```python
merge["age_group"] = pd.cut(merge["Age"], bins=[0,15, 30], labels=["Young", "Adult"])
merge
```
